```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
```

```python
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

```python
fig.canvas.get_supported_filetypes_grouped()
```

```python
fig.savefig('line.pdf')
```

```python
# and this opens the SVG
fig.savefig('line.svg')
#webbrowser.open('line.svg')
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
```

```python
plt.scatter(x, y);
```

```python
plt.plot(x, y, '-ok', color='black');
```

```python
rng = np.random.RandomState(0)
plt.scatter(x, y, c=rng.randn(30), cmap='summer');
```

```python
data = rng.randn(1000)
plt.hist(data)
```

```python
plt.hist(data,
         bins=20,
         normed=True,
         alpha=0.5,
         color='steelblue');
```

```python
np.histogram(data, bins=5)
```

```python
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')
```

```python
fig, ax = plt.subplots(2, 3, sharex='all', sharey='all')
ax[0, 0].plot(x)
ax[0, 1].plot(y)
ax[0, 2].plot(np.cos(x))
ax[0, 2].plot(np.sin(x))
ax[1, 0].plot(np.linspace(0, 1, 20))
ax[1, 1].plot(rng.randn(20))
ax[1, 2].plot(rng.randint(0, 20, 20))
```
